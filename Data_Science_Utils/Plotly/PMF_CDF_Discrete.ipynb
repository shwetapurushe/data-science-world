```python
import plotly
import plotly.graph_objects as go
import plotly.io as pio
import plotly.figure_factory as ff

pio.templates.default = 'plotly_dark'
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
import numpy as np 
```

```python
n_defaults = np.random.binomial(100, 0.05, 10000)
```

CDF

```python
x, y = ecdf(n_defaults)
```

```python
fig = go.Figure(
    data=go.Scatter(x=x, y=y, mode='lines'),
    layout_title_text='CDF with a Scatter plot',
    layout_xaxis_title ="Number of Defaults",
    layout_yaxis_title= "ECDF"
)

fig.show()
```

```python
fig = go.Figure(
    data= go.Scattergl(x= x, y = y, mode='markers'),
    layout_title_text = "Plotting a CDF with ScatterGL"
)
fig.show()
```
